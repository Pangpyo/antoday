```python
import time
import requests
from bs4 import BeautifulSoup

# 초기 페이지와 페이지 수 설정
base_url = "https://finance.naver.com/news/mainnews.naver"
pages_to_scrape = 5  # 원하는 기사 개수
articles = []

for page in range(1, pages_to_scrape + 1):
    # 각 페이지의 URL 설정
    url = f"{base_url}?page={page}"

    # 요청 시작
    r = requests.get(url)

    # 뉴스 페이지 파싱
    soup = BeautifulSoup(r.text, "html.parser")

    # 각 언론사별 카드형 박스 수집 - 제목, 본문 미리보기
    news_boxes = soup.find_all("ul", "newsList")

    # 기사 제목과 링크 수집
    main_news_titles = []
    main_news_links = []

    for news_box in news_boxes:
        article_list = news_box.find("li", "block1").find_all("dl")
        for arti in article_list:
            content = arti.find("dd", "articleSubject")
            if content:
                main_news_titles.append(content.find("a").text.strip())
                main_news_links.append(content.find("a")["href"])

    # 각 페이지의 기사를 수집
    for title, link in zip(main_news_titles, main_news_links):
        article_url = f"https://finance.naver.com{link}"
        r = requests.get(article_url)
        time.sleep(0.5)

        # 기사 원문 파싱
        soup = BeautifulSoup(r.text, "html.parser")
        article_body = soup.find("div", id="content")
        if article_body:
            article_body = article_body.text.strip()
        else:
            article_body = '기사 내용 못 찾음'

        # 파싱한 원문을 articles에 삽입
        articles.append(
            {
                "title": title,
                "content": article_body
            }
        )
```
